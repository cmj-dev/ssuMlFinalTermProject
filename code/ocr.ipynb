```python
import os
from google.cloud import vision
from tqdm import tqdm
```

```python
def detect_response(path):
    """Detects text in the file."""
    from google.cloud import vision
    client = vision.ImageAnnotatorClient()

    with open(path, 'rb') as image_file:
        content = image_file.read()

    image = vision.Image(content=content)

    response = client.text_detection(image=image)
    #texts = response.text
    if response.error.message:
        raise Exception(
            '{}\nFor more info on error messages, check: '
            'https://cloud.google.com/apis/design/errors'.format(
                response.error.message))
    
    return response
```

```python
textList = []
blockInfoList = []
for i in tqdm(range(948)):
    response = detect_response(f"/Users/minjae/dev/23_1/mlProjectFinal/data/images/{i}.jpg")
    text = response.text_annotations[0].description
    blockList = response.full_text_annotation.pages[0].blocks
    blockInfo = []
    for block in blockList:
        x = 0
        y = 0
        l = len(block.bounding_box.vertices)
        for vertice in block.bounding_box.vertices:
            x += vertice.x
            y += vertice.y
        blockInfo.append([x/l, y/l])
    textList.append(text)
    blockInfoList.append(blockInfo)
```

```python
import pandas as pd
```

```python
pd.DataFrame({'text':textList,'blockInfo':blockInfoList}).to_csv('ocrInfo.csv')
```
